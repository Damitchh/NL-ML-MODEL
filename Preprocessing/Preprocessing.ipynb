```python
#Load Data

import pandas as pd

df = pd.read_csv('/Users/muhammadrezapulungan/Documents/Skripsi/peduli_lindungi_labelfix2.csv', sep=';')
print(df)
```

```python
df['label'].value_counts()
```

```python
#remove Number
df.review = df.review.str.replace('\d+', '')
df.head()
```

```python
#remove punctuation
import string
def remove_punctuation(df): return df.translate(str.maketrans("","",string.punctuation ))
df['review'] = df['review'].apply(remove_punctuation)
df.head()
```

```python
#remove single char #regular Expression
import re
def remove_single_char(df): return re.sub(r"\b[a-zA-Z]\b", "", df) 
df['review'] = df['review'].apply(remove_single_char)
df.head()
```

```python
#remove whitespace
import re
def remove_whitespace_multiple(review):
    return re.sub('\s+',' ',review)
df['review'] = df['review'].apply(remove_whitespace_multiple)
df.head()
```

```python
#casefolding
review = []
for index, row in df.iterrows():
    review.append(row["review"].lower())
    
df["review"] = review
df.head()
```

```python
# stemming dengan sastrawi
review = []
for index, row in df.iterrows():
    review.append(stemmer.stem(row["review"]))
    
df["review"] = review
df.head()
```

```python
#Tokenisasi
from nltk.tokenize import word_tokenize
def review_tokenizing(review):
    return word_tokenize(review)
df['review_tokens'] = df['review'].apply(review_tokenizing)
df.head()
```

```python
#stopword removal pake NLTK
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')
list_stopwords = stopwords.words('indonesian')
list_stopwords.extend(['yg', 'hayuck' , 'nya', 'kt', 'jg' ,
                       'krn', 'pke','donk','aj','saya','gua','kn',
                       'in','oh','si','nih','utk','aja','n','u','y','k'])
txt_stopword = pd.read_csv('/Users/muhammadrezapulungan/Documents/Skripsi/stopwords.txt', names= ["stopwords"], header = None)
list_stopwords.extend(txt_stopword["stopwords"][0].split(' '))

list_stopwords = set(list_stopwords)

def stopwords_removal(words):
    return [word for word in words if word not in list_stopwords]

df['sw_clean'] = df['review_tokens'].apply(stopwords_removal)
df.head()
```

```python
#Normalisasi
normalize_word = pd.read_excel('/Users/muhammadrezapulungan/Documents/Skripsi/normalisasi.xlsx')
normalize_word_dict = {}
for index, row in normalize_word.iterrows():
  if row[0] not in normalize_word_dict:
      normalize_word_dict[row[0]] = row[1] 

def normalized_term(document):
  return [normalize_word_dict[term] 
          if term in normalize_word_dict 
          else term for term in document]

df['normalisasi_clean'] = df['sw_clean'].apply(normalized_term)
df.head()
```
